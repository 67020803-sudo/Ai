```python
# ตัวอย่างที่ 4.1
import numpy as np

# 1. เตรียมข้อมูล (Inputs & Outputs)
# ข้อมูลนำเข้า: [A, B]
inputs = np.array([[0, 0],
                   [1, 1],
                   [1, 0],
                   [0, 1]])

# ผลลัพธ์ที่คาดหวัง (Target): 0 OR 0 = 0, 1 OR 1 = 1, เป็นต้น
outputs = np.array([[0, 1, 1, 1]]).T

# 2. ตั้งค่าเริ่มต้น
np.random.seed(1)
# สุ่ม Weight เริ่มต้น 2 ค่า สำหรับ 2 Input (ช่วง -1 ถึง 1)
weights = 2 * np.random.random((2, 1)) - 1

# 3. ฟังก์ชัน Sigmoid (Activation Function)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 4. การฝึกฝน (Training Loop)
print("กำลังเริ่มฝึกฝน...")
for iteration in range(50000):
    # Forward Propagation: คำนวณผลลัพธ์ปัจจุบัน
    output = sigmoid(np.dot(inputs, weights))

    # คำนวณ Error (ความผิดพลาด)
    error = outputs - output

    # Backpropagation: ปรับค่า Weight ตาม Error ที่เกิดขึ้น
    # สูตร: input * error * (derivative of sigmoid)
    adjustment = np.dot(inputs.T, error * output * (1 - output))
    weights += adjustment

print("ฝึกฝนเสร็จสิ้น!")
print("-" * 30)

# 5. การทดสอบ (Testing)
print("ค่าน้ำหนัก (Weights) หลังฝึกฝน:")
print(weights)
print("-" * 30)

# 6) การทดสอบ (Testing)
test_cases = [
    np.array([0, 0]),  # OR ควรได้ 0
    np.array([1, 1]),  # OR ควรได้ 1
    np.array([1, 0]),  # OR ควรได้ 1
    np.array([0, 1])   # OR ควรได้ 1
]

print("การทดสอบผลลัพธ์ Logical OR:")
for test_data in test_cases:
    prediction = sigmoid(np.dot(test_data, weights))
    print(f"Input {test_data}: Raw={prediction[0]:.4f}, Rounded={int(round(prediction[0]))}")
```

```python
import numpy as np

# 1. เตรียมข้อมูล (Inputs & Outputs) - เปลี่ยนเป็น Logical AND
# ข้อมูลนำเข้า: [A, B]
inputs = np.array([[0, 0],
                   [1, 1],
                   [1, 0],
                   [0, 1]])

# ผลลัพธ์ที่คาดหวัง (Target): A AND B = 0, 1, 0, 0
outputs = np.array([[0, 1, 0, 0]]).T

# 2. ตั้งค่าเริ่มต้น
np.random.seed(1)
# สุ่ม Weight เริ่มต้น 2 ค่า สำหรับ 2 Input (ช่วง -1 ถึง 1)
weights = 2 * np.random.random((2, 1)) - 1

# 3. ฟังก์ชัน Sigmoid (Activation Function)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 4. การฝึกฝน (Training Loop)
print("กำลังเริ่มฝึกฝน Logical AND...")
for iteration in range(50000):
    # Forward Propagation: คำนวณผลลัพธ์ปัจจุบัน
    output = sigmoid(np.dot(inputs, weights))

    # คำนวณ Error (ความผิดพลาด)
    error = outputs - output

    # Backpropagation: ปรับค่า Weight ตาม Error
    adjustment = np.dot(inputs.T, error * output * (1 - output))
    weights += adjustment

print("ฝึกฝนเสร็จสิ้น!")
print("-" * 30)

# 5. การทดสอบ (Testing)
print("ค่าน้ำหนัก (Weights) หลังฝึกฝน:")
print(weights)
print("-" * 30)

# ทดลองทำนายค่าใหม่ๆ
test_cases = [
    np.array([0, 0]),  # ควรได้ 0
    np.array([1, 1]),  # ควรได้ 1
    np.array([1, 0]),  # ควรได้ 0
    np.array([0, 1])   # ควรได้ 0
]

print("การทดสอบผลลัพธ์ Logical AND:")
for i, test_data in enumerate(test_cases):
    prediction = sigmoid(np.dot(test_data, weights))
    print(f"Input {test_data}: Raw={prediction[0]:.4f}, Rounded={round(prediction[0])}")
```

```python
# Example 4.2 Modified - 3 Inputs Multi-Layer Perceptron
# ตรรกะ: A OR B OR C (OR ของ 3 ตัว)
from sklearn.neural_network import MLPClassifier

# 3 Inputs: [A, B, C] และผลลัพธ์ A OR B OR C
X = [
    [0., 0., 0.],  # 0 OR 0 OR 0 = 0
    [1., 1., 1.],  # 1 OR 1 OR 1 = 1
    [0., 1., 0.],  # 0 OR 1 OR 0 = 1
    [1., 0., 0.],  # 1 OR 0 OR 0 = 1
    [0., 0., 1.],  # 0 OR 0 OR 1 = 1
    [1., 0., 1.],  # 1 OR 0 OR 1 = 1
    [0., 1., 1.],  # 0 OR 1 OR 1 = 1
    [1., 1., 0.]   # 1 OR 1 OR 0 = 1
]

y = [0, 1, 1, 1, 1, 1, 1, 1]  # ผลลัพธ์ OR ของ 3 ตัว

# สร้าง MLP (ปรับ hidden_layer_sizes ให้เหมาะกับ 3 inputs)
clf = MLPClassifier(
    solver='lbfgs',
    alpha=1e-5,
    hidden_layer_sizes=(6, 3),  # เพิ่ม neurons ให้เหมาะกับ 3 inputs
    random_state=1,
    max_iter=1000  # เพิ่ม iterations
)

clf.fit(X, y)

print("=== ผลการทดสอบ ===")
print("ทดสอบ input [2., 2., 2.]:", clf.predict([[2., 2., 2.]]))
print("ทดสอบ input [-1., -2., -3.]:", clf.predict([[-1., -2., -3.]]))

print("\n=== โครงสร้าง Weights ===")
print([coef.shape for coef in clf.coefs_])

print("\n=== Accuracy ===")
print("Training Accuracy:", clf.score(X, y))
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# โหลดข้อมูล
whole_data = load_breast_cancer()
X_data = whole_data.data
y_data = whole_data.target

# แบ่งข้อมูล
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.3, random_state=7
)

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
print(f"Features: {X_train.shape[1]}")

# สร้าง MLP Model (3 layers)
model = Sequential([
    Dense(10, activation='relu', input_shape=(X_train.shape[1],), name='hidden1'),
    Dense(10, activation='relu', name='hidden2'),
    Dense(1, activation='sigmoid', name='output')  # Sigmoid สำหรับ binary classification
])

model.summary()

# Compile ใหม่ - เหมาะกับ Classification
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',      # ✅ เปลี่ยนเป็น BCE
    metrics=['accuracy']             # ✅ เปลี่ยนเป็น accuracy
)

# ฝึกโมเดล
history = model.fit(
    X_train, y_train,
    batch_size=50,
    validation_split=0.2,
    epochs=100,
    verbose=1
)

# ทดสอบ
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\n=== ผลลัพธ์ ===")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.1f}%)")

# Prediction
y_pred = (model.predict(X_test) > 0.5).astype(int).flatten()
print(f"\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import time
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# โหลดและเตรียมข้อมูล (เหมือนเดิม)
whole_data = load_breast_cancer()
X_data, y_data = whole_data.data, whole_data.target
X_train, X_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.3, random_state=7
)

print("ข้อมูล: Train={}, Test={}".format(X_train.shape[0], X_test.shape[0]))
print("=" * 60)

# ฟังก์ชันสร้างและทดสอบโมเดล
def train_and_test(hidden_neurons, model_name):
    start_time = time.time()

    # สร้างโมเดล
    model = Sequential([
        Dense(hidden_neurons, activation='relu', input_shape=(30,), name='hidden1'),
        Dense(hidden_neurons, activation='relu', name='hidden2'),
        Dense(1, activation='sigmoid', name='output')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # ฝึกโมเดล (ลด epochs เหลือ 50 เพื่อเทียบ training time)
    history = model.fit(X_train, y_train, batch_size=50, validation_split=0.2,
                       epochs=50, verbose=0)

    training_time = time.time() - start_time

    # ทดสอบ
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    y_pred = (model.predict(X_test, verbose=0) > 0.5).astype(int).flatten()

    print(f"📊 {model_name} ({hidden_neurons} neurons)")
    print(f"   Training Time: {training_time:.2f} วินาที")
    print(f"   Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"   Test Loss: {test_loss:.4f}")
    print("-" * 40)

    return test_accuracy, training_time, history # Return history object as well

# รันทั้ง 2 โมเดล
print("🔄 เปรียบเทียบ 10 vs 5 Neurons (50 epochs)")
print()

acc_10, time_10, history_10 = train_and_test(10, "โมเดลเดิม") # Capture history_10
acc_5, time_5, history_5 = train_and_test(5, "โมเดลใหม่")   # Capture history_5

# สรุปเปรียบเทียบ
print("=" * 60)
print("📈 สรุปเปรียบเทียบ")
print(f"{'':<15} {'10 Neurons':<12} {'5 Neurons':<12}")
print(f"{'Training Time':<15} {time_10:<12.2f}s {time_5:<12.2f}s "
      f"(Diff: {(time_10 - time_5):+.2f}s)")
print(f"{'Test Accuracy':<15} {acc_10:<12.4f} {acc_5:<12.4f} "
      f"(Diff: {((acc_10 - acc_5)*100):+.2f}%)")
print(f"{'Total Parameters':<15}", end=" ")
model_10_params = Sequential([Dense(10, input_shape=(30,)), Dense(10), Dense(1)])
model_5_params = Sequential([Dense(5, input_shape=(30,)), Dense(5), Dense(1)])
print(f"{model_10_params.count_params():<12} {model_5_params.count_params():<12}")

# เพิ่มโค้ดนี้เพื่อพล็อตกราฟ (ถ้าต้องการ)
import matplotlib.pyplot as plt

# Use the captured history objects directly
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history_10.history['loss'], label='10 Neurons')
plt.plot(history_5.history['loss'], label='5 Neurons')
plt.title('Training Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_10.history['val_loss'], label='10 Neurons')
plt.plot(history_5.history['val_loss'], label='5 Neurons')
plt.title('Validation Loss')
plt.legend()
plt.show()
```
