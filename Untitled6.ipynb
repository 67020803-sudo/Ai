```python
#Example 3.1 Python SVM Classifications
from sklearn import svm

# X: ข้อมูลคุณลักษณะ (Features)
# [[ความสูง (ซม.), น้ำหนัก (กก.), ขนาดรองเท้า (UK)]]
X = [[170, 70, 10], [180, 80, 12], [170, 65, 8], [160, 55, 7]]

# y: ป้ายกำกับ (Labels) หรือ คลาส (Classes) ที่ทราบแล้ว
# 0: ชาย (Male), 1: หญิง (Female)
y = [0, 0, 1, 1]

# 1. สร้างตัวจำแนก (Classifier)
clf = svm.SVC()

# 2. ฝึกฝนโมเดล (Training)
clf.fit(X, y)

# 3. ทำการทำนาย (Prediction) สำหรับข้อมูลตัวอย่างใหม่
p = clf.predict([[160, 60, 7]])

# 4. แสดงผลลัพธ์
print(p)
```

```python
# EXERCISE 3.2 - แก้ไขจาก Example 3.2
# ใช้ Petal length และ Petal width เป็นคุณลักษณะ

from sklearn import svm, datasets

# 1. โหลดชุดข้อมูล Iris
iris = datasets.load_iris()

# X: เลือกคุณลักษณะที่สามและสี่ (Petal length และ Petal width)
# ดัชนี 2: Petal length, ดัชนี 3: Petal width
X = iris.data[:, 2:]

# y: ป้ายกำกับเป้าหมาย (0: Setosa, 1: Versicolour, 2:Virginica)
y = iris.target

print(f"ป้ายกำกับทั้งหมดในชุดข้อมูล (y): \n{y}\n")

# 2. สร้างและฝึกฝนตัวจำแนก (Classifier Training)
clf = svm.SVC()
clf.fit(X, y)

# 3. กำหนดข้อมูลตัวอย่างใหม่สำหรับการทำนาย
# ใช้ค่าจาก Example 3.2 แต่แทนด้วย Petal length และ width
# เนื่องจากไม่มีค่า Sepal ในโมเดลนี้ เราจึงใช้ค่าสมมติของ Petal (เช่น Petal length=4.0, Petal width=1.3)
petal_sample = [[4.0, 1.3]]

print(f"ข้อมูลตัวอย่างสำหรับการทำนาย (Petal length, Petal width): {petal_sample}")

# 4. ทำนายดอกไม้
p = clf.predict(petal_sample)

print(f"ผลลัพธ์การทำนาย: {p}")
```

```python
#Example 3.3 Python SVM Iris CSV Classifications
from sklearn import svm, datasets
import pandas as pd

# 0 . กำหนดตำแหน่งไฟล์ CSV ที่มีข้อมูล Iris จาก GitHub
csv_url = 'https://raw.githubusercontent.com/thammarat-ai/prinAI6-28/refs/heads/main/chapter3/iris.csv'

# csv_peng ='https://raw.githubusercontent.com/thammarat-ai/prinAI6-28/refs/heads/main/chapter3/penguins.csv'

# 1. อ่านข้อมูลจากไฟล์ CSV ด้วย Pandas
# กำหนดชื่อคอลัมน์เอง และข้ามแถวแรก (skiprows=1)
df = pd.read_csv(csv_url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'], skiprows=1)

# 2. X: เลือกคุณลักษณะ 2 ตัวแรก (Sepal length, Sepal width)
# .values เพื่อแปลง DataFrame เป็น NumPy array
X = df.values[:,:2]

# 3. เตรียมป้ายกำกับ (Labels)
s = df['species'] # ดึงคอลัมน์ชื่อสปีชีส์ (ข้อความ)
d = dict([(y,x) for x,y in enumerate(sorted(set(s)))]) # สร้าง Dictionary เพื่อแปลงชื่อสปีชีส์เป็นตัวเลข (0, 1, 2)
y = [d[x] for x in s] # แปลงคอลัมน์ species เป็นป้ายกำกับตัวเลข (y)

# 4. สร้างและฝึกฝนตัวจำแนก (Classifier Training)
clf = svm.SVC()
clf.fit(X, y)

# 5. ทำนาย
p = clf.predict([[5.4, 3.2]])
print(p)
```

```python
# EXERCISE 3.3 - แก้ไขจาก Example 3.4
# แสดงผลกราฟกระจาย (Scatter Plot) ของ Sepal Length และ Sepal Width

from sklearn import svm, datasets
import pandas as pd
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
import numpy as np # เพิ่ม NumPy เพื่อจัดการกับข้อมูล

# 0. โหลดข้อมูลจาก URL
csv_url = 'https://raw.githubusercontent.com/thammarat-ai/prinAI6-28/refs/heads/main/chapter3/iris.csv'
df = pd.read_csv(csv_url, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'], skiprows=1)

# การจัดการข้อมูลสูญหาย
df = df.dropna()

# 1. เตรียมป้ายกำกับตัวเลข (Numerical Labels) สำหรับสีในกราฟ
s = df['species']
# สร้าง Dictionary เพื่อแปลงชื่อสปีชีส์เป็นตัวเลข (0, 1, 2)
d = dict([(y,x) for x,y in enumerate(sorted(set(s)))])
y = np.array([d[x] for x in s]) # แปลงเป็น NumPy array สำหรับใช้เป็นสี (c)

# 2. เตรียมคุณลักษณะ (Features) สำหรับแกน X และ Y
# Sepal Length
sepal_length = df['sepal_length']
# Sepal Width
sepal_width = df['sepal_width']

# --- ส่วนการพล็อตกราฟกระจาย (Scatter Plot) ---
## 3. สร้างกราฟกระจาย
pyplot.figure(figsize=(8, 6)) # กำหนดขนาดของกราฟ

# c=y ใช้ป้ายกำกับตัวเลขเป็นสี เพื่อแยกประเภทของสปีชีส์
# cmap='viridis' กำหนดชุดสี
scatter = pyplot.scatter(sepal_length, sepal_width, c=y, cmap='viridis')

# 4. เพิ่มคำอธิบายกราฟ
pyplot.xlabel('Sepal Length (cm)')
pyplot.ylabel('Sepal Width (cm)')
pyplot.title('Iris Dataset: Sepal Length vs Sepal Width (Scatter Plot)')

# สร้างคำอธิบายสัญลักษณ์ (Legend)
legend1 = pyplot.legend(*scatter.legend_elements(), title="Species (Class)")
pyplot.gca().add_artist(legend1)

# 5. แสดงกราฟ
pyplot.show()

# --- ส่วน SVM (ยังคงอยู่ตามโจทย์เดิม) ---
X = df.values[:,:2]
clf = svm.SVC()
clf.fit(X, y)
p = clf.predict([[5.4, 3.2]])
print(f"\nผลลัพธ์การทำนายด้วย SVM: {p}")
```

```python
# EXERCISE 3.4 - Python SVM Breast Cancer Classification + Histogram
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from matplotlib import pyplot # นำเข้า pyplot สำหรับการสร้างกราฟ

# 1. โหลดชุดข้อมูล
cancer = load_breast_cancer()
X = cancer.data # คุณลักษณะทั้งหมด (Features)
y = cancer.target # ป้ายกำกับทั้งหมด (Labels)

# 2. การวิเคราะห์ข้อมูลเบื้องต้น (EDA): สร้าง Histogram
# แปลงข้อมูลคุณลักษณะ (X) และชื่อคุณลักษณะ (feature_names) เป็น DataFrame
df = pd.DataFrame(X, columns=cancer.feature_names)

# เลือกคุณลักษณะที่ต้องการพล็อต (mean radius, mean texture, mean smoothness, mean perimeter)
# Note: ใช้ mean perimeter แทน 'size' และ 'mean smoothness' แทน 'smoothness'
features_to_plot = ['mean radius', 'mean texture', 'mean smoothness', 'mean perimeter']

# สร้าง Histogram
df[features_to_plot].hist(figsize=(10, 8))
pyplot.suptitle('Histogram of Selected Breast Cancer Features', y=1.02, fontsize=16)
pyplot.tight_layout()
pyplot.show()
# -----------------------------------------------------------------

# 3. แบ่งข้อมูล (Split Data)
# 80% สำหรับ Training, 20% สำหรับ Testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)

# 4. ฝึกฝนโมเดล SVM
clf = SVC()
clf.fit(X_train, y_train)

# 5. ทำนายผล
y_predict = clf.predict(X_test)

# 6. แสดง Confusion Matrix และ Classification Report
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm,
                         index=['is_benign (1)', 'is_malignant (0)'],
                         columns=['predicted_benign','predicted_malignant'])

print("\n--- Confusion Matrix ---")
print(confusion)

print("\n--- Classification Report ---")
print(classification_report(y_test, y_predict))
```

```python
# EXERCISE 3.5 - Naive Bayes with Model Serialization/Deserialization

from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
# ต้องติดตั้ง joblib ก่อน: pip install joblib
import joblib

# 1. โหลดข้อมูล
X, y = load_iris(return_X_y=True)

# 2. ฝึกฝนโมเดล Naive Bayes
print("--- 1. Training Model ---")
clf = GaussianNB()
clf.fit(X, y)
print("Model training complete.")

# 3. บันทึกโมเดลลงไฟล์ (Serialization)
model_filename = 'naive_bayes_iris_model.pkl'
joblib.dump(clf, model_filename)
print(f"Model saved to file: {model_filename}")

# 4. โหลดโมเดลจากไฟล์ (Deserialization)
print("\n--- 2. Loading Model ---")
# clf2 คือโมเดลที่โหลดกลับเข้ามา
clf2 = joblib.load(model_filename)
print("Model loaded successfully.")

# 5. ทำนายด้วยโมเดลที่โหลดมา
sample_data = [[5.0, 3.4, 1.5, 0.4]]
p = clf2.predict(sample_data)

print("\n--- 3. Prediction ---")
print(f"Input sample: {sample_data}")
print(f"Prediction result (Class): {p}")
# ผลลัพธ์ที่ได้จะเป็น [0] (Iris Setosa) เนื่องจากคุณลักษณะนี้อยู่ในช่วงของ Setosa
```

```python
# EXERCISE 3.6 - LDA Classification with 6 Features and 2000 Samples

from sklearn.datasets import make_classification
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# 1. สร้างชุดข้อมูลจำลองใหม่
# n_samples = 2000
# n_features = 6
X, y = make_classification(n_samples=2000,
                           n_features=6,
                           n_informative=2,
                           n_redundant=0,
                           random_state=0,
                           shuffle=False)

print(f"ขนาดชุดข้อมูล (Samples, Features): {X.shape}")
# แสดงข้อมูลคุณลักษณะ (X)
print("ข้อมูลคุณลักษณะ (X):")
print(X)

# 2. สร้างและฝึกฝนโมเดล LDA
clf = LinearDiscriminantAnalysis()
clf.fit(X, y)

# 3. ทำนายผลสำหรับตัวอย่างใหม่ (ต้องใช้ 6 ค่าเนื่องจากมี 6 features)
p = clf.predict([[0, 0, 0, 0, 0, 0]])

print("\nผลลัพธ์การทำนายสำหรับตัวอย่าง [0, 0, 0, 0, 0, 0]:")
print(p)
```

```python
# EXERCISE 3.7 - PCA with Breast Cancer Dataset

import numpy as np
import matplotlib.pyplot as plt
from sklearn import decomposition
from sklearn.datasets import load_breast_cancer # โหลดชุดข้อมูลมะเร็งเต้านม

# 1. โหลดชุดข้อมูลมะเร็งเต้านม
cancer = load_breast_cancer()
X = cancer.data   # คุณลักษณะทั้งหมด (30 features)
y = cancer.target # ป้ายกำกับ (0: malignant, 1: benign)

# 2. พล็อตข้อมูลเดิม (ใช้คุณลักษณะ 2 ตัวแรกสำหรับการแสดงผล)
f = plt.figure(1, figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.get_cmap('viridis', 2))
plt.xlabel(cancer.feature_names[0]) # 'mean radius'
plt.ylabel(cancer.feature_names[1]) # 'mean texture'
plt.title('Original Breast Cancer Data (Features 1 & 2)')
plt.colorbar(ticks=[0, 1], label='Target (0: Malignant, 1: Benign)')
f.show()

# 3. ทำ PCA (ลดมิติจาก 30 เหลือ 3 องค์ประกอบ)
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X1 = pca.transform(X) # ข้อมูลถูกแปลงเป็น 3 มิติใหม่

# 4. พล็อตข้อมูล PCA (ใช้ 2 องค์ประกอบแรกสำหรับการแสดงผล)
g = plt.figure(2, figsize=(10, 5))
plt.scatter(X1[:, 0], X1[:, 1], c=y, cmap=plt.cm.get_cmap('viridis', 2))
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA Transformed Data (PCA1 vs PCA2)')
plt.colorbar(ticks=[0, 1], label='Target (0: Malignant, 1: Benign)')
g.show()
```

```python
# EXERCISE 3.8 - Decision Tree Classification on Wine Data

from sklearn.datasets import load_wine # ต้องเปลี่ยนไปโหลดชุดข้อมูลไวน์
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score # เพิ่มการวัดความแม่นยำ (Accuracy)

# 1. โหลดชุดข้อมูลไวน์
wine = load_wine()
X = wine.data     # คุณลักษณะทั้งหมด (เช่น alcohol, malic acid, etc.)
y = wine.target   # ป้ายกำกับ (คลาสของไวน์ 3 ประเภท)

# 2. แบ่งข้อมูล (Split Data)
# ใช้ 30% เป็นชุดทดสอบ (เป็นค่ามาตรฐานที่นิยมใช้)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. สร้างและฝึกฝนโมเดล Decision Tree
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# 4. ทำนายผล
y_pred = clf.predict(X_test)

# 5. คำนวณและแสดงผลประสิทธิภาพ (ใช้ทั้งวิธีเดิมและวิธี Accuracy Score)
N = y_test.shape[0] # จำนวนตัวอย่างรวมในชุดทดสอบ
C = (y_test == y_pred).sum() # จำนวนตัวอย่างที่ทำนายถูกต้อง

print("--- Wine Data Decision Tree Classification ---")
print(f"Total points (N): {N}")
print(f"Correctly labeled points (C): {C}")

# แสดงความแม่นยำ (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.4f}")
```

```python
# EXERCISE 3.9 - Random Forest Classification on Diabetes Data

from sklearn.datasets import load_diabetes # ต้องเปลี่ยนไปโหลดชุดข้อมูลเบาหวาน
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np # นำเข้า NumPy สำหรับการจัดการข้อมูล

# 1. โหลดชุดข้อมูลเบาหวาน
# Note: ชุดข้อมูลเบาหวานมักใช้สำหรับการถดถอย (Regression) แต่สามารถใช้กับ Random Forest Classifier ได้
# เราจะใช้ข้อมูลตัวอย่างเพื่อสาธิตการเปลี่ยนชุดข้อมูล
diabetes = load_diabetes()
X = diabetes.data     # คุณลักษณะทั้งหมด (เช่น BMI, Blood Pressure, etc.)
y = diabetes.target   # ป้ายกำกับเป้าหมาย (ค่าความคืบหน้าของโรค)

# เนื่องจาก RandomForestClassifier ต้องการป้ายกำกับแบบ Discrete (Classification),
# แต่ชุดข้อมูล Diabetes มีป้ายกำกับแบบต่อเนื่อง (Continuous)
# เราจะทำการแปลงปัญหาให้เป็นปัญหา "จำแนกประเภท" อย่างง่าย โดยแบ่งค่า y ออกเป็น 2 คลาส:
# ค่าน้อยกว่าค่ามัธยฐานเป็น 0, ค่ามากกว่าเท่ากับค่ามัธยฐานเป็น 1
median_y = np.median(y)
y_classified = (y >= median_y).astype(int)

# 2. แบ่งข้อมูล (Split Data)
X_train, X_test, y_train, y_test = train_test_split(X, y_classified, test_size=0.3, random_state=42)

# 3. สร้างและฝึกฝนโมเดล Random Forest
# ใช้ n_estimators เพื่อระบุจำนวนต้นไม้ในป่า
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# 4. ทำนายผล
y_pred = clf.predict(X_test)

# 5. คำนวณและแสดงผลประสิทธิภาพ
N = y_test.shape[0]
C = (y_test == y_pred).sum()

print("--- Diabetes Data Random Forest Classification ---")
print(f"Total points (N): {N}")
print(f"Correctly labeled points (C): {C}")
print(f"Accuracy: {C/N:.4f}")
```

```python
#Example 3.13 K-Nearest Neighbors Classification
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

clf = KNeighborsClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("Total points: %d Correctly labeled points : %d" %(y_test.shape[0],(y_test == y_pred).sum()))
```

```python
#Example 3.14 Modified: Comparison of Different Regression Models
# using the Diabetes dataset

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR # Support Vector Regression (SVR) แทน SVC
from sklearn.linear_model import LinearRegression # Linear Regression แทน Naive Bayes
from sklearn.tree import DecisionTreeRegressor # Decision Tree Regressor แทน Classifier
from sklearn.ensemble import RandomForestRegressor # Random Forest Regressor แทน Classifier
from sklearn.neighbors import KNeighborsRegressor # K-Nearest Neighbors Regressor แทน Classifier
from sklearn.neural_network import MLPRegressor # Neural Network Regressor แทน Classifier
from sklearn.metrics import mean_squared_error # ใช้ MSE สำหรับการประเมินผลงาน Regression

# เนื่องจาก Diabetes เป็น Regression เราจึงใช้โมเดล Regression แทน Classification

names = [
    "SVR",
    "Linear Regression",
    "Decision Tree Regressor",
    "Random Forest Regressor",
    "K-Neighbors Regressor",
    "Neural Network Regressor"
]

classifiers = [
    SVR(),
    LinearRegression(),
    DecisionTreeRegressor(random_state=0),
    RandomForestRegressor(random_state=0),
    KNeighborsRegressor(),
    MLPRegressor(alpha=1, max_iter=1000, random_state=0)
]

# 1. โหลดข้อมูล Diabetes แทน Iris
X, y = load_diabetes(return_X_y=True)

# 2. แบ่งข้อมูล (ใช้สัดส่วนเดิม 50%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

print("Comparison of Regression Models on Diabetes Data:")
print("--------------------------------------------------")
# 3. วนลูปฝึกสอนโมเดล, คาดการณ์, และประเมินผล
# สำหรับ Regression เราจะใช้ Mean Squared Error (MSE) โดยค่าที่ "น้อยที่สุด" คือดีที่สุด
for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    # คำนวณ Mean Squared Error (MSE)
    mse = mean_squared_error(y_test, y_pred)

    # เนื่องจาก clf.score() ของ Regressor จะคืนค่า R-squared (R2)
    # ซึ่งค่าที่ "มากที่สุด" คือดีที่สุด เราจะพิมพ์ทั้งสองค่าเพื่อเปรียบเทียบ
    r2_score = clf.score(X_test, y_test)

    print(f"{name}:")
    print(f"  R-squared (R2): {r2_score:.4f} (ยิ่งสูงยิ่งดี)")
    print(f"  Mean Squared Error (MSE): {mse:.2f} (ยิ่งต่ำยิ่งดี)\n")
```
